```python
import pymysql

import time
from datetime import datetime
from operator import attrgetter

import matplotlib.pyplot as plt

# Connection
con = pymysql.connect(host='220.149.31.104',
                      user='lora',
                      password='loralora',
                      db='loratest',
                      charset='utf8')

cursorObject = con.cursor()

print("connect successful!!")
print()

sqlQuery = "select port, sequence, date from dummydata where port ="


# Execute the sqlQuery
for i in range(4) :
    port = i + 10
    cursorObject.execute(sqlQuery+str(port)+";")

    # Fetch all the rows
    rows = cursorObject.fetchall()
    print("port number : ",port)
        
    print("packet 개수: ", len(rows))
        
    print("Block Error Rate : ",(rows[len(rows)-1][1] - (len(rows) - rows[0][1])) / len(rows) * 100, "%")
    

    # 전체 전송 시간
    timeStampFirst = int(round(rows[0][2].timestamp()))
    timeStampLast = int(round(rows[len(rows)-1][2].timestamp()))
    second = timeStampLast - timeStampFirst
    
    print("전송시간 : ",second, "sec")
    
    print("Data Rate", (rows[len(rows)-1][1]) / second)

    print()
# data rate 구하는 방법
# (마지막 sequence - 처음 sequence) * packet의 bit 수/(마지막 row의 date - 첫번째 row의 date)

# 실제 데이터에서는 port별로, data rate가 다르다고 가정,



# plt.plot()
# plt.show()

# Close connection
con.commit()
con.close()
```
